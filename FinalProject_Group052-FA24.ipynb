## Setup

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Baseball Players Dataset

The Baseball Players dataset was cleaned and reformatted for analysis. We renamed columns for better readability, removed rows with missing values, and eliminated duplicates. The `Name` column, originally containing both first and last names separated by an underscore, was split into `first_name` and `last_name`, which were then combined into a new column, `last_name,_first_name`, in the format "LastName, FirstName" for consistency. Finally, we retained only the relevant columns: `last_name,_first_name`, `Team`, `Position`, `Height_inches`, `Weight_pounds`, and `Age`.

```python
baseball_players = pd.read_csv('baseball_players.csv')
baseball_players. rename(columns={ 'Height(inches)': 'Height_inches', 'Weight(pounds)': 'Weight_pounds'}, inplace=True)
baseball_players.dropna(inplace=True)
baseball_players. drop_duplicates(inplace=True)
baseball_players[['first_name', 'last_name']] = baseball_players['Name'].str.split('_', n=1, expand=True)
baseball_players['last_name,_first_name'] = baseball_players['last_name'] + ', ' + baseball_players['first_name']
baseball_players = baseball_players[['last_name,_first_name', 'Team', 'Position', 'Height_inches', 'Weight_pounds', 'Age']]

baseball_players
```

## Exit Velocity Dataset

The Exit Velocity dataset was cleaned and prepared for analysis. Column names were standardized by replacing spaces with underscores and converting them to lowercase for consistency. Rows with missing values were removed to ensure complete data, and duplicates were eliminated. Finally, only relevant columns were included, `last_name,_first_name`, `player_id`, `avg_hit_speed`, `max_hit_speed`, `brl_percent`, and `ev95percent`, which are key offensive performance metrics.

```python
exit_velocity = pd.read_csv('exit_velocity.csv')
exit_velocity.columns = exit_velocity.columns.str.replace(' ', '_').str.lower()
exit_velocity.dropna(inplace=True)
exit_velocity.drop_duplicates(inplace=True)
exit_velocity = exit_velocity[['last_name,_first_name', 'player_id', 'avg_hit_speed', 'max_hit_speed', 'brl_percent', 'ev95percent']]
exit_velocity
```

## Outs Above Average Dataset

The Outs Above Average dataset was cleaned and prepared for analysis. Column names were standardized by replacing spaces with underscores and converting them to lowercase for consistency. Rows with missing values were removed to ensure completeness, and duplicate rows were eliminated. Only relevant defensive performance metrics were retained, including `last_name,_first_name`, `player_id`, `outs_above_average`, and `fielding_runs_prevented`.

```python
outs_above_average = pd.read_csv('outs_above_average.csv')
outs_above_average.columns = outs_above_average.columns.str.replace(' ', '_').str.lower()
outs_above_average.dropna(inplace=True)
outs_above_average.drop_duplicates(inplace=True)
outs_above_average = outs_above_average[['last_name,_first_name', 'player_id', 'outs_above_average', 'fielding_runs_prevented']]
outs_above_average
```

## Stats Dataset

The Stats dataset was cleaned and prepared for analysis. Column names were standardized by replacing spaces with underscores and converting them to lowercase for consistency. Rows with missing values were removed to ensure complete data, and duplicate rows were eliminated. Only the relevant batting statistics were retained, including `last_name,_first_name`, `player_id`, `pa`, `k_percent`, `bb_percent`, `xba`, and `hard_hit_percent`.

```python
stats = pd.read_csv('stats.csv')
stats.columns = stats.columns.str.replace(' ', '_').str.lower()
stats.dropna(inplace=True)
stats.drop_duplicates(inplace=True)
stats = stats[['last_name,_first_name', 'player_id', 'pa', 'k_percent', 'bb_percent', 'xba', 'hard_hit_percent']]
stats
```

## Merge

The datasets were merged to create a comprehensive dataset for analysis. First, the **Exit Velocity** dataset was merged with the **Stats** dataset using `player_id` and `last_name,_first_name` as keys. Next, the resulting dataset was merged with the **Outs Above Average** dataset, again using `player_id` and `last_name,_first_name`. Finally, this merged dataset was joined with the **Baseball Players** dataset on `last_name,_first_name` to incorporate player attributes such as height, weight, and age. 

After merging, we cleaned the final dataset by dropping duplicates and rows with missing values. We then created two new columns: 
1. **Age**: Rounded down using `np.floor` to ensure age values were integers.
2. **BMI**: Calculated using the formula, providing a measure of players' body mass relative to their height.

This final dataset is ready for analysis, linking physical attributes to both offensive and defensive performance metrics.

```python
merged_data1 = pd.merge(exit_velocity, stats, on=['player_id', 'last_name,_first_name'], how='inner')

merged_data2 = pd.merge(merged_data1, outs_above_average, on=['player_id', 'last_name,_first_name'], how='inner')

final_merged_data = pd.merge(merged_data2, baseball_players, left_on='last_name,_first_name', right_on='last_name,_first_name', how='inner')
final_merged_data.dropna(inplace=True)
final_merged_data. drop_duplicates(inplace=True)
final_merged_data['Age'] = np.floor(final_merged_data['Age'])
final_merged_data['BMI'] = (final_merged_data['Weight_pounds'] * 703) / (final_merged_data['Height_inches'] ** 2)

final_merged_data
```

### Exploring the BMI Distribution Among Various Players

This section analyzes the distribution of BMI among various players. BMI is the combination of one's height and weight in order to give insights on an individual's body composition. Through examination of the BMI among various MLB players, we aim to uncover any patterns that may connect to a player's performance.

We provide two visualizations in order to understand the data:

1. A **Histogram** in order to show the frequency of players across multiple BMI ranges.
2. A **Line Plot** in order to highlight any trends in BMI distribution.

```python
bmi_counts = final_merged_data['BMI'].value_counts().reset_index()
bmi_counts.columns = ['BMI', 'Count']
bmi_counts = bmi_counts.sort_values(by='BMI')

#histogram
plt.figure(figsize=(12, 6))
sns.histplot(data=bmi_counts, x='BMI',bins=15)
plt.title('Distribution of Players by BMI', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Count of Players', fontsize=12)

#line plot
plt.figure(figsize=(12, 6))
sns.lineplot(data=bmi_counts, x='BMI', y='Count')
plt.title('Distribution of Players by BMI', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Count of Players', fontsize=12)
```

The histogram and line plot trends suggest that most players have a BMI around 25, which aligns with the general expectation that players typically fall within a range that supports both agility and power. This BMI distribution may reflect the balance between agility (important for defensive positions) and power (important for offensive roles). With this in mind, the negative correlation predicted between BMI and defensive statistics is supported by the idea that possessing a higher BMI may reduce agility, limiting a player's effectiveness in positions requiring speed and other quick movements. The positive correlation between BMI and offensive statistics aligns with the notion that physically larger players, who are more powerful, are better suited for offensive roles such as batters where power is essential.

### Visualizing BMI Density
This section further explores the distribution of BMI values among players utilizing a Kernel Density Estimate (KDE) plot. The purpose of this plot is to offer us deeper insights into the density and spread of BMI values within the dataset.

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=final_merged_data, x="BMI", fill=True, color="blue", alpha=0.6)
plt.axvline(final_merged_data["BMI"].mean(), color="red", linestyle="--")
plt.title("Distribution of BMI (KDE Plot)", fontsize=16)
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Density", fontsize=12)
```

The **red line** in the graph represents its mean, serving as a reference point for the dataset's central value. As shown in the graph, the KDE plot indicates a peak at around a BMI of 25, which suggests that it's the average BMI of the players in the dataset. Our hypothesis involves a positive correlation between BMI and offense, suggesting that players with a BMI of around 25 may have better offensive statistics. There exists a steep drop in the amount of players that possess a BMI that is greater than 25, indicating that players tend to stay at or below a BMI of 25 for their athletic performance.

### Defensive Metrics and BMI

This section will focus on analyzing the connection between a player's BMI and their defensive capabilities. The metrics we will use are a player's **Outs Above Average** and **Fielding Runs Prevented**. We'll be using regression plots and a correlation matrix in order to identify any patterns or trends that may support or challenge our hypothesis.

```python
#we are looking at BMI in relation to defense capabilities
plt.figure(figsize=(8, 6))
sns.regplot(data=final_merged_data, x='BMI', y='outs_above_average', line_kws={'color': 'red'})
plt.title('BMI vs. Outs Above Average', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Outs Above Average', fontsize=12)
plt.grid(True)

plt.figure(figsize=(8, 6))
sns.regplot(data=final_merged_data, x='BMI', y='fielding_runs_prevented', line_kws={'color': 'red'})
plt.title('BMI vs. Fielding Runs Prevented', fontsize=16)
plt.xlabel('BMI', fontsize=12)
plt.ylabel('Fielding Runs Prevented', fontsize=12)
plt.grid(True)

#correlation analysis
defensive_metrics = final_merged_data[["BMI", "outs_above_average", "fielding_runs_prevented"]]
correlation_matrix_defense = defensive_metrics.corr()
print("Correlation Matrix for Defense Metrics:")
print(correlation_matrix_defense)
bmi_defense_correlations = correlation_matrix_defense.loc["BMI", ["outs_above_average", "fielding_runs_prevented"]]
print("\nCorrelation of BMI with Defensive Metrics:")
print(bmi_defense_correlations)
```

The regression plots examining BMI vs. Outs Above Average and BMI vs. Fielding Runs Prevented show that there is a weak negative relationsip between BMI and our defensive metrics. The regression lines indicate a slight downward trend, where players with higher BMIs tend to have lower defensive effectiveness. The plots showcase that there isn't really a strong relationship between BMI and defensive performance, as the lines are weak slopes. Our correlation matrix further reinforces our findings, as the correlation between BMI and Outs Above Average is -0.150 and BMI and Fielding Runs Prevented is at -0.137. This data suggests taht a higher BMI is weakly connected to lower defensive performance. 

### BMI and Offensive Performance

This section analyzes how a player's BMI correlates to key offensive performance metrics such as **Average Hit Speed**, **Max Hit Speed**, and **Expected Batting Average**. Through data visualizations and a correlation matrix, we aim to determine if a player's BMI plays a significant role in offensive success.

```python

#BMI vs. Average Hit Speed
plt.figure(figsize=(8, 6))
sns.regplot(data=final_merged_data, x="BMI", y="avg_hit_speed",  line_kws={'color': 'red'})
plt.title("BMI vs. Average Hit Speed", fontsize=16)
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Average Hit Speed", fontsize=12)
plt.grid(alpha=0.3)

#BMI vs. Max Hit Speed
plt.figure(figsize=(8, 6))
sns.regplot(data=final_merged_data, x="BMI", y="max_hit_speed", line_kws={'color': 'red'})
plt.title("BMI vs. Max Hit Speed", fontsize=16)
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Max Hit Speed", fontsize=12)
plt.grid(alpha=0.3)

#BMI vs. xBA
plt.figure(figsize=(8, 6))
sns.regplot(data=final_merged_data, x="BMI", y="xba", scatter_kws={'alpha': 0.6}, line_kws={'color': 'red'})
plt.title("BMI vs. Expected Batting Average (xBA)", fontsize=16)
plt.xlabel("BMI", fontsize=12)
plt.ylabel("Expected Batting Average (xBA)", fontsize=12)
plt.grid(alpha=0.3)

offensive_metrics = final_merged_data[["BMI", "avg_hit_speed", "max_hit_speed", "xba"]]
correlation_matrix = offensive_metrics.corr()
print("Correlation Matrix:")
print(correlation_matrix)
bmi_correlations = correlation_matrix.loc["BMI", ["avg_hit_speed", "max_hit_speed", "xba"]]
print("\nCorrelation of BMI with Offensive Metrics:")
print(bmi_correlations)
```

The correlation matrix along with our regression plots can be seen as partially supporting our hypothesis. The correlation between BMI and offensive metrics average hit speed is (0.14), max hit speed is (0.13), and expected batting average is (0.05). These results may align with our hypothesis are they are highlight a positive relationship between BMI and offensive performance, but the correlations aren't strong enough, indicating that BMI alone isn't a strong predictor of offensive success. Our regression plots further support this statement, as although there is a slight positive trend between BMI and our offensive metrics, the weak slopes suggest that BMI is only one factor among many that influence offensive performance.
